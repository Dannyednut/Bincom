```python
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import pandas as pd
```

```python

datasets = fetch_california_housing()
X = pd.DataFrame(datasets.data, columns=datasets.feature_names)
y = pd.DataFrame(datasets.target, columns=['value'])
X.head()
```

```python
y.head()
```

```python
X.isnull().sum()
```

```python
y.isnull().sum()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x= X['MedInc'], y=y['value'], s=2, c='black')
plt.xlabel('MedInc')
plt.ylabel('Value')
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
scaler = StandardScaler()
model = LinearRegression()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, predictions)
```

```python
r2 = r2_score(y_test, predictions)
```

```python
print(f'mean_squared_error: {mse:.2f}', )
print(f'r2_score: {r2:.2f}')
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(y_test, predictions, label='Predictions', s=5, c='black')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', label='Perfect Prediction')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Regression Plot')
plt.legend()
plt.show()
```
